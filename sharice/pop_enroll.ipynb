```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

import seaborn as sns

file = "Resources/UK&SPPoSc.csv"

data =pd.read_csv(file)

data.head()
```

```python
yr_10 = "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011"
plt.plot(yr_10, data.iloc[0, 2:13], label="UK_pop_", marker='o')
plt.plot(yr_10, data.iloc[1, 2:13], label="UK_sci_enroll", marker='o')
plt.plot(yr_10, data.iloc[2, 2:13], label="SP_pop", marker='o')
plt.plot(yr_10, data.iloc[3, 2:13], label="SP_sci_enroll", marker='o')

plt.title("UK vs Spain Population of the official age for tertiary education, both sexes​ & Percentage of students in tertiary education enrolled in Science programmes, both sexes")
plt.xlabel("years")
plt.ylabel("Percentage")
plt.legend()
```

```python
fig,ax = plt.subplots()
ax.plot(yr_10, data.iloc[0, 2:13], label="UK_pop_", marker='o', color='b')
ax.plot(yr_10, data.iloc[2, 2:13], label="SP_pop", marker='o', color='r')
ax.set_xlabel("years")
#ax.set_ylabel("UK vs Spain Population of the official age for tertiary education, both sexes")
ax2 = ax.twinx()
ax2.plot(yr_10, data.iloc[1, 2:13], label="UK_sci_enroll", marker='v', color='b')
ax2.plot(yr_10, data.iloc[3, 2:13], label="SP_sci_enroll", marker='v', color='r')
ax2.set_ylabel("Unemployment, total (% of total labor force)")
plt.title("UK vs Spain Population of the official age for tertiary education, both sexes​ & Percentage of students in tertiary education enrolled in Science programmes, both sexes")
legend_elements = [Line2D([0], [0],marker='o', color='b', lw=1, label="UK govt exp"), Line2D([0], [0],marker='o', color='r', lw=1, label="UK & SP Population"),
                   Line2D([0], [0],marker='v', color='b', lw=1, label="UK unemployment"), Line2D([0], [0],marker='v', color='r', lw=1, label="UK & SP Science Enroll")]
ax.legend(handles=legend_elements)
plt.show()
```
