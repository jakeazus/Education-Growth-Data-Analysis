```python
file_to_load = "/Users/Chibbbins/desktop/data-analysis-project-1/chandler/Resources/EdStatsData.csv"


raw_education_data = pd.read_csv(file_to_load)

raw_education_data.head()
```

```python
raw_education_data_UK = raw_education_data.loc[(raw_education_data['Country Name'] == "United Kingdom")]

ind_pull = ["Government expenditure in pre-primary institutions as % of GDP (%)",
"Government expenditure in primary institutions as % of GDP (%)",
"Government expenditure in secondary institutions education as % of GDP (%)",
"Government expenditure on education as % of GDP (%)",
"Government expenditure on post-secondary non-tertiary education as % of GDP (%)",
"Government expenditure on pre-primary education as % of GDP (%)",
"Government expenditure on primary education as % of GDP (%)",
"Government expenditure on secondary and post-secondary non-tertiary vocational education as % of GDP (%)",
"Government expenditure on secondary education as % of GDP (%)",
"Government expenditure on tertiary education as % of GDP (%)",
"Government expenditure in educational institutions as % of GDP (%)"]

clean_UK = pd.DataFrame()

for x in ind_pull:

    clean_UK = clean_UK.append(raw_education_data_UK.loc[raw_education_data_UK['Indicator Name'] == x])     

finished_UK = clean_UK.drop(columns=["2000", "2012", "2013", "2014", "2015","2016", "2017", "2020", "2025",
                                     "2030", "2035", "2040", "2045", "2050", "2055", "2060", "2065", "2070", "2075",
                                     "2080", "2085", "2090", "2095", "2100", "Country Code", "Indicator Code"])
finished_UK.head()
```

```python
from matplotlib.lines import Line2D

fig,ax = plt.subplots()
ax.plot(ref_yr, finished_UK.iloc[3, 2:13], marker='o', color='b')
ax.plot(ref_yr, finished_SP.iloc[3, 2:13], marker='o', color='r')
ax.set_xlabel("years")
ax.set_ylabel("Government expenditure as % of GDP")
ax2 = ax.twinx()
ax2.plot(ref_yr, finished_compare_UK.iloc[0, 2:13], marker='v', color='b')
ax2.plot(ref_yr, finished_compare_SP.iloc[0, 2:13], marker='v', color='r')
ax2.set_ylabel("Unemployment, total (% of total labor force)")
plt.title("Government expenditure on education as % of GDP and Unemployment, total (% of total labor force) over 10 yr period")



legend_elements = [Line2D([0], [0],marker='o', color='b', lw=1, label="UK govt exp"), Line2D([0], [0],marker='o', color='r', lw=1, label="SP govt exp"),
                   Line2D([0], [0],marker='v', color='b', lw=1, label="UK unemployment"), Line2D([0], [0],marker='v', color='r', lw=1, label="SP unemployment")]
ax.legend(handles=legend_elements)
plt.show()
```
